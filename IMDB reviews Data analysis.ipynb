```python
import pandas as pd
```

```python
df = pd.read_csv('IMDB-Dataset.csv')
print(df.head())
```

```python
df.columns
```

```python
#Cleaning the Text Data
# Before generating the word cloud, we need to clean the text data which involves:

# 1. Removing punctuation

# 2. Converting text to lowercase

# 3. Removing stopwords i.e common words like "the", "is", "and"
```

```python
import re
from wordcloud import STOPWORDS
text = ' '.join(df['review'].astype(str).tolist())#converting all review entries into single text
text = re.sub(r'[^A-Za-z\s]', '', text)#no alphabatic characters are relaced by ''
text = text.lower()
stopwords = set(STOPWORDS)
text = ' '.join(word for word in text.split() if word not in stopwords)
print(text[:100])
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

wordcloud=WordCloud(width=700,height=300,background_color='white').generate(text)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  
plt.title("IMDB Movie Reviews Word Cloud")
plt.show()
```
